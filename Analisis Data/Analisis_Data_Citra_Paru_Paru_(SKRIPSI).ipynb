#FUNCTION  UNTUK MENGHITUNG NILAI MEAN RGB DAN STANDART DEVIASI RGB

- Letakkan Folder Yang Berisikan Kelas Data Gambar
- Jalakan Function
- Hasil Akan tersimpan ke dalam bentuk Excel


```python
import os
import pandas as pd
from PIL import Image
import numpy as np

# Path ke folder gambar
folder_path = "/content/drive/MyDrive/@SKRIPSI/Analisis Data/DATA PARU-PARU (PRIMER)/TUBERCULOSIS"

result_df = pd.DataFrame(columns=["Gambar", "Mean R", "Mean G", "Mean B", "Std Deviasi R", "Std Deviasi G", "Std Deviasi B", "Std Deviasi Keseluruhan"])

for filename in os.listdir(folder_path):
    if filename.endswith(".jpg") or filename.endswith(".png"):
        image_path = os.path.join(folder_path, filename)
        image = Image.open(image_path)

        # Periksa jika gambar dalam mode grayscale, konversi menjadi mode RGB
        if image.mode == "L":
            image = image.convert("RGB")

        image_array = np.array(image)

        mean_r = np.mean(image_array[:, :, 0])
        mean_g = np.mean(image_array[:, :, 1])
        mean_b = np.mean(image_array[:, :, 2])
        std_dev_r = np.std(image_array[:, :, 0])
        std_dev_g = np.std(image_array[:, :, 1])
        std_dev_b = np.std(image_array[:, :, 2])
        std_dev = np.std(image_array)

        result_df = result_df.append({"Gambar": filename, "Mean R": round(mean_r, 2),
                                      "Mean G": round(mean_g, 2), "Mean B": round(mean_b, 2),
                                      "Std Deviasi R": round(std_dev_r, 2),
                                      "Std Deviasi G": round(std_dev_g, 2),
                                      "Std Deviasi B": round(std_dev_b, 2),
                                      "Std Deviasi Keseluruhan": round(std_dev, 2)},
                                     ignore_index=True)

result_df.to_excel("TUBERCULOSIS.xlsx", index=False)
```

```python

import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_excel("/content/drive/MyDrive/@SKRIPSI/Analisis Data/BRONKITIS.xlsx")

mean_r = df["Mean R"].mean()
mean_g = df['Mean G'].mean()
mean_b = df['Mean B'].mean()
mean_std_r = df['Std Deviasi R'].mean()
mean_std_g = df['Std Deviasi G'].mean()
mean_std_b = df['Std Deviasi B'].mean()
mean_std_all = df['Std Deviasi Keseluruhan'].mean()

# Menampilkan hasil
print("\nMean R:")
print(mean_r)
print("\nMean G:")
print(mean_g)
print("\nMean B:")
print(mean_b)
print("\nMean std R:")
print(mean_std_r)
print("\nMean std G:")
print(mean_std_g)
print("\nMean std B:")
print(mean_std_b)
print("\nMean std All:")
print(mean_std_all)



mean_values = [mean_r, mean_g, mean_b, mean_std_r, mean_std_g, mean_std_b, mean_std_all]

# Label untuk histogram
labels = ['Mean R', 'Mean G', 'Mean B', 'Std Deviasi R', 'Std Deviasi G', 'Std Deviasi B', 'Std Deviasi Keseluruhan']

# Membuat plot histogram
plt.figure(figsize=(10, 6))
plt.bar(labels, mean_values)
plt.xlabel('Variabel')
plt.ylabel('Rata-rata')
plt.title('Analisis Data Bronkitis')
plt.xticks(rotation=45)

# Menampilkan plot histogram
plt.show()
```
